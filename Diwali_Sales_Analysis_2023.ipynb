## Exploratory Data Analysis

```python
df.columns
```

```python
# To see who are top 5 customers based on number of orders
a=df.groupby(["Cust_name"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(5)
sns.set(rc={"figure.figsize":(8,5)})
sns.barplot(data=a,x="Cust_name",y="Orders")
```

```python
# To see top 5 products popular among customers
a=df.groupby(["Product_ID"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(5)
print(a)
```

```python
# To see number of Male and Female customers
a=sns.countplot(x="Gender", data=df)
for bars in a.containers:
    a.bar_label(bars)
# We have more female customers more than twice compare to males
```

```python
# To see occupation of our customers
a=sns.countplot(x="Occupation", data=df)
for bars in a.containers:
    a.bar_label(bars)
sns.set(rc={"figure.figsize":(28,5)})
# Majority of our customers are from IT, Healthcare and Aviation sector
```

```python
# Analyze customers based on their age group
a=sns.countplot(x="Age Group", data=df)
for bars in a.containers:
    a.bar_label(bars)
# Majority of customers are from age 26 to 45 means Working Adults are our mass customers
```

```python
# Marital Status of Customers
a=sns.countplot(x="Marital_Status", data=df)
sns.set(rc={"figure.figsize":(5,4)})
for bars in a.containers:
    a.bar_label(bars)
```

```python
# Orders made state wise
a=sns.countplot(x="State", data=df)
for bars in a.containers:
    a.bar_label(bars)
sns.set(rc={"figure.figsize":(28,5)})
# U.P, Maharastra,Karnataka are top 3 states
```

```python
# Zone wise orders
a=sns.countplot(x="Zone", data=df)
for bars in a.containers:
    a.bar_label(bars)
# Maximum Orders from Central Zone
```

```python
# Popular product categories among customers
a=sns.countplot(x="Product_Category", data=df)
for bars in a.containers:
    a.bar_label(bars)
sns.set(rc={"figure.figsize":(28,5)})
# Clothing & Apparel, Food and Electronics & Gadgets are top 3 product categories
```

```python
# Gender wise sales amount
sale_gender=df.groupby(["Gender"],as_index=False)["Amount"].sum()
sns.barplot(x="Gender",y="Amount",data=sale_gender)
# females are purchasing goods of more amounts then men
```

```python
# Order count Age group wise comparison between male and female
a=sns.countplot(x="Age Group",data=df,hue="Gender")
for bars in a.containers:
    a.bar_label(bars)
# female of age group 26-35 are most frequent buyers 
```

```python
# Sales Age Group wise
sale_age=df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Age Group",y="Amount",data=sale_age)
```

```python
# Total orders from top 10 states
sales_state=df.groupby(["State"], as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
sns.set(rc={"figure.figsize":(28,5)})
sns.barplot(x="State",y="Orders",data=sales_state)
# UP, Maharastra, Karnataka are top 3 states making maximum orders
```

```python
# Total Sales from top 10 states
sales_state=df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.barplot(x="State",y="Amount",data=sales_state)
# Sales is same as orders but Harayana, Bihar and Gujarat has more purchasing power than Kerala
```

```python
# Married Status
a=sns.countplot(data=df,x="Marital_Status")
sns.set(rc={"figure.figsize":(5,5)})
for bars in a.containers:
    a.bar_label(bars)
```
